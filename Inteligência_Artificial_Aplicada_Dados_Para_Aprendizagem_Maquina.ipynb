#Machine Learning

**Machine Learning ou apredizagem em máquina é o campo de estudo de programas de computadores que melhoram seu desempenho de acordo com sua experiência.**


Seu foco é permitir que agentes artificiais melhorem os resultados de medida de desempenho de acordo com dados “inputados” por humanos ou capturados ao longo da sua interação com o ambiente, sem a necessidade de programá-lo explicitamente para aqueles resultados.​ Também conhecido como aprendizado indutivo: aprender uma regra geral a partir de exemplos.



**APRENDIZAGEM SUPERVISIONADA**

Na aprendizagem de máquina é aplica técnicas, procedimentos e algoritmos que resultam na “aprendizagem” das regras baseadas no conjunto de dados apresentado como entrada. É possível realizar modelos para diferentes problemas, como: classificação, predição, ordenação e tomada de decisão.

A qualidade dos resultados está diretamente relacionada à qualidade dos dados apresentados aos algoritmos. Assim como em ciências de dados, é comum as etapas de definição do problema, exploração e preparação dos dados. A prática de aprendizado de máquina envolve uma fase de treinamento, seguida por uma fase de teste. Independentemente do algoritmo, o algoritmo de aprendizado de máquina é treinado em um conjunto de dados.


O objetivo do algoritmo é fazer julgamentos sobre os dados de teste com base nos recursos que ele "aprendeu" com os dados de treinamento.


**Aprendizagem Supervisionada -** o treinamento é realizado por meio de dados “rotulados”, a base de dados contém, além dos atributos (variáveis), a variável “Alvo (target)”. É uma aprendizagem por meio de exemplos.


A aprendizagem supervisionada abrange dois tipos de problemas, classificação e regressão.


**Problemas de classificação** são aqueles em que buscamos um resultado dentro de classe ou categoria, como: se tentando classificar uma imagem como cavalo ou humano, possuímos apenas duas classes possíveis.

Os algoritmos utilizados são: KNN, SVN, árvore de decisão, regressão logística, entre outros.

**Regressão** são problemas em que o nosso resultado pode assumir valores contínuos. Por exemplo, quando o objetivo é realizar a predição do valor de um imóvel, a variável valor do imóvel contém valores contínuos de preço. Os algoritmos utilizados são: KNN, Regressão linear, Regressão polinomial, entre outros.


**APREDIZAGEM NÃO SUPERVISIONADA**

O atributo alvo (target) não existe ou não é fornecido durante a etapa de aprendizagem, os dados não estão rotulados. Neste caso, os dados são agrupados ou clusterizados em categorias definidas por similaridade. Muito utilizada para encontrar padrões nos dados em análise exploratória e redução de dimensionalidade. Os algoritmos utilizados são: KMeans, DBScan, PCA, entre outros.


**APREDIZAGEM POR REFORÇO**

É muito útil em jogos, sistemas robóticos, carros autônomos entre outros que interagem com o ambiente, pois envolve o aprendizado de ações ou comportamentos com base em reforços positivos ou negativos recebidos pelo agente, trata-se de um processo iterativo constante. Neste tipo de aprendizagem, o agente não possui conjunto de dados inicialmente, isso é feito de forma iterativa recebendo recompensas ou penalidades.

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
from sklearn.model_selection import train_test_split

# Carregar o California Housing Dataset
housing_data = fetch_california_housing()

# Criar DataFrame
df = pd.DataFrame(housing_data.data, columns=housing_data.feature_names)
df['Target'] = housing_data.target

# Usar dois atributos para predizer o valor de saída
X = df[['MedInc', 'AveRooms']]
Y = df['Target']

# Verificar o formato das tabelas de dados e saídas
print(f"Formato das tabelas de dados {X.shape} e saídas {Y.shape}")

# Separar 20% dos dados para o teste
X_treino, X_teste, Y_treino, Y_teste = train_test_split(X, Y, test_size=0.2, random_state=42)

# Visualizar as primeiras linhas dos dados de treino e teste
print(X_treino.head())
print(Y_treino.head())
```

```python
# Importa as bibliotecas
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Cria o modelo de regressão linear
lin_model = LinearRegression()

# Treina o modelo de machine learning
lin_model.fit(X_treino, Y_treino)

# Realiza previsões no conjunto de teste
y_teste_predito = lin_model.predict(X_teste)

# Calcula o Erro Quadrático Médio (MSE)
mse = mean_squared_error(Y_teste, y_teste_predito)
print("Erro Quadrático Médio (MSE): %.2f" % mse)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np

from sklearn.datasets import make_blobs

# Geração aleatória de conjunto de dados no formato de bolhas
# 300 amostras
# 4 clusters
# 0,6 de desvio padrão para cada clusters

X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```
